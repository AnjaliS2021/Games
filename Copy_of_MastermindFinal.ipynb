```python
<!DOCTYPE html>  
<html>  
<head>  
<title> JavaScript MasterMind </title>  
</head>
<body>
  <canvas id="MastermindCanvasId" width="700" height="500"></canvas>
  <p id="MastermindParaId">Let the game begin!!</p>
<script>
 const canvas = document.getElementById("MastermindCanvasId");
  const ctx = canvas.getContext("2d");
  const para = document.getElementById("MastermindParaId");
  function WriteToPara(text) {
    //para.innerText = para.innerText + "\n" + text;
    para.innerText = text + "\n" + para.innerText;
  }

  class Hole {
    constructor(x, y, radius, color) {
      this.x = x;
      this.y = y;
      this.radius = radius;
      this.defaultColor = color;
      this.color = color;
    }
    toString() {
      return '[object Hole]';
    }
    draw() {
      ctx.beginPath();
      ctx.fillStyle = this.color;
      // x, y, radiusX, radiusY, rotation, startAngle, endAngle
      ctx.ellipse(this.x, this.y, this.radius, this.radius, 0, 0, Math.PI*2);
      ctx.fill();
      ctx.closePath();
      ctx.stroke();
    }
    highlight(color) {
      if (color != null) {
        this.color = color;
        this.draw();
      }
      var radius = this.radius - 2;
      ctx.beginPath();
      // x, y, radiusX, radiusY, rotation, startAngle, endAngle
      ctx.ellipse(this.x, this.y, radius, radius, 0, 0, Math.PI*2);
      ctx.closePath();
      ctx.stroke();
    }
    clearHighlight() {
      this.draw();
    }
    changeColor(color) {
      this.color = color;
      this.draw();
    }
    isClicked(x, y) {
      var withinX = x >= (this.x - this.radius) && x <= (this.x + this.radius);
      var withinY = y >= (this.y - this.radius) && y <= (this.y + this.radius);
      return withinX && withinY;
    }
  }
  class Square {
    constructor(x, y, color) {
      this.x = x;
      this.y = y;
      this.side = 24;
      this.color = color;
    }
    toString() {
      return '[object Square]';
    }
    draw() {
      ctx.beginPath();
      ctx.fillStyle = this.color;
      ctx.fillRect(this.x, this.y, this.side, this.side);
      ctx.closePath();
      ctx.stroke();
    }
    isClicked(x, y) {
      var withinX = x >= this.x && x <= (this.x + this.side);
      var withinY = y >= this.y && y <= (this.y + this.side);
      return withinX && withinY;
    }
    highlight() {
      ctx.beginPath();
      var diff = 2;
      ctx.rect(
        this.x + diff, this.y + diff, this.side - 2*diff, this.side - 2*diff);
      ctx.closePath();
      ctx.stroke();
    }
    clearHighlight() {
      this.draw();
    }
  }
  class Column {
    constructor(x, y, color) {
      this.x = x;
      this.y = y;
      this.color = color;
      this.holes = [];
      this.clickedHole = null;
      this.prevClickedHole = null;
    }
    drawRect() {
      ctx.beginPath();
      ctx.fillStyle = this.color;
      ctx.fillRect(this.x, this.y, 30, 150);
      ctx.closePath();
      ctx.stroke();
    }
    drawHoles() {
      const numHoles = 4;
      const radius = 10;
      const space = 15;
      for (var i = 0; i < numHoles; i++) {
        const x = this.x + 5 + radius;
        const y = this.y + 10 + radius + i*3*radius;
        const hole = new Hole(x, y, radius, "LightGray");
        hole.draw();
        this.holes.push(hole);
      }
    }
    changeHoleColors(colors) {
      for (var i = 0; i < colors.length; i++) {
        var color = colors[i];
        var hole = this.holes[i];
        hole.changeColor(color);
      }
    }
    draw() {
      this.drawRect();
      this.drawHoles();
    }
    holeClicked(x, y) {
      for (var i = 0; i < this.holes.length; i++) {
        var hole = this.holes[i];
        if (hole.isClicked(x, y)) {
          this.prevClickedHole = this.clickedHole;
          this.clickedHole = hole;
          // WriteToPara('found clicked hole index: ' + i.toString());
          return true;
        }
      }
      return false;
    }
    hightlightHole(color) {
      if (this.clickedHole != null) {
        this.clickedHole.highlight(color);
      }
      if (this.prevClickedHole != null) {
        this.prevClickedHole.clearHighlight();
      }
    }
    getHoleColors() {
      var colors = [];
      for (var i = 0; i < this.holes.length; i++) {
        var hole = this.holes[i];
        if (hole.color === hole.defaultColor) {
          colors.push(null);
        }
        else {
          colors.push(hole.color);
        }
      }
      return colors;
    }
  }
  class PinsColumn extends Column{
    constructor(x, y, color) {
      super(x, y, color);
      this.radius = 4;
      this.space = 5;
      this.isClicked = false;
    }
    drawRect() {}
    drawHoles() {
      const numHoles = 4;
      for (var i = 0; i < 2; i++) {
        for (var j = 0; j < 2; j++) {
          const x = this.x + this.space + this.radius + i*3*this.radius;
          const y = this.y + 2*this.space + this.radius + j*3*this.radius;
          const hole = new Hole(x, y, this.radius, "RosyBrown");
          hole.draw();
          this.holes.push(hole);
        }
      }
    }
    changePinsColors(bulls, cows) {
      var colors = [];
      for (var i = 0; i < bulls; i++) {
        colors.push('red');
      }
      for (var j = 0; j < cows; j++) {
        colors.push('white');
      }
      for (var i = 0; i < 4-cows-bulls; i++) {
        colors.push('RosyBrown');
      }
      this.changeHoleColors(colors);
    }
    pinsClicked(x, y) {
      var widthX = 3*this.space + 5*this.radius;
      var widthY = 4*this.space + 5*this.radius;
      var withinX = x >= this.x && x <= (this.x + widthX);
      var withinY = y >= this.y && y <= (this.y + widthY);
      this.isClicked = withinX && withinY;
    }
  }
  class ColorsPicker {
    constructor(x, y) {
      this.x = x;
      this.y = y;
      //green, pink, 
      this.colors = ['Chartreuse', 'Violet', 'BlueViolet', 'MediumTurquoise', 'FireBrick', 'DarkOrange'];
      this.squares = [];
      this.clickedSquare = null;
      this.prevClickedSquare = null;
    }
    draw() {
      const num = this.colors.length;
      for (var i = 0; i < num; i++) {
        var color = this.colors[i];
        var x = this.x + 17;
        var y = this.y + 15 + i*32;
        var square = new Square(x, y, color);
        square.draw();
        this.squares.push(square);
      }
    }
    colorClicked(x, y) {
      for (var i = 0; i < this.squares.length; i++) {
        var square = this.squares[i];
        if (square.isClicked(x, y)) {
          this.prevClickedSquare = this.clickedSquare;
          this.clickedSquare = square;
          // WriteToPara('found clicked square: ' + square.color);
          return true;
        }
      }
      return false;
    }
    onClickEvent(x, y) {
      var clicked = this.colorClicked(x, y);
      if (!clicked) {
        return false;
      }
      if (this.prevClickedSquare == this.clickedSquare) {
        this.clickedSquare = null;
      }
      if (this.clickedSquare != null) {
        this.clickedSquare.highlight();
      }
      if (this.prevClickedSquare != null) {
        this.prevClickedSquare.clearHighlight(); 
        this.prevClickedSquare = null;       
      }
      return true;
    }
  }
  class Logo {
    constructor(x, y) {
      this.x = x;
      this.y = y;
    }
    draw() {
      var x = this.x + 170;
      var y = this.y + 250;
      ctx.beginPath();
      ctx.font = "bold 45px arial";
      ctx.fillStyle = "white"
      ctx.fillText("MasterMind", x, y);
      ctx.closePath();
      ctx.stroke();
    }
  }
  class CodeMaker {
    constructor(ansCol, colorPicker) {
      this.ansCol = ansCol;
      this.colorPicker = colorPicker;
      this.codeIndexes = [];
      this.colors = [];
    }
    pickColorIndexes() {
      for (var round = 0; round < 4; round++) {
        var index = Math.floor(Math.random() * this.colorPicker.colors.length);
        this.codeIndexes.push(index);
      }
    }
    pickColors() {
      for (var i = 0; i < this.codeIndexes.length; i++) {
        var index = this.codeIndexes[i];
        var color = this.colorPicker.colors[index];
        this.colors.push(color);
      }
    } 
    run() {
      this.pickColorIndexes();
      this.pickColors();
    } 
  }
  class CodeBreaker {
    constructor(cols, ansCol, pinsCols, colorsPicker, ansColors) {
      this.cols = cols;
      this.ansCol = ansCol;
      this.pinsCols = pinsCols;
      this.ansColors = ansColors;
      this.colorsPicker = colorsPicker;
      this.activeColIndex = 0;
    }
    pickActiveHole(x, y) {
      var col = this.cols[this.activeColIndex];
      var clicked = col.holeClicked(x, y);
      if (!clicked) {
        return false;
      }
      var color = null;
      if (this.colorsPicker.clickedSquare != null) {
        color = this.colorsPicker.clickedSquare.color;
      }
      // WriteToPara('color: ' + color);
      col.hightlightHole(color);
      return true;
    }
    pickActivePinsCol(x, y) {
      var pinsCol = this.pinsCols[this.activeColIndex];
      pinsCol.pinsClicked(x, y);
      //WriteToPara('pins clicked: ' + pinsCol.isClicked.toString());
      return pinsCol.isClicked;
    }
    copyArray(arr) {
      var copyArr = [];
      for (var i = 0; i < arr.length; i++) {
        copyArr.push(arr[i]);
      }
      return copyArr;
    }
    scoreBulls(colColors, ansColors) {
      var bulls = 0;
      for (var i = 0; i < colColors.length; i++) {
        if (colColors[i] != null && ansColors[i] != null
           && colColors[i] === ansColors[i]) {
          bulls++;
          colColors[i] = null;
          ansColors[i] = null;
        }
      }
      return bulls;
    }
    scoreCows(colColors, ansColors) {
      var cows = 0;
      for (var i = 0; i < colColors.length; i++) {
        for (var j = 0; j < ansColors.length; j++) {
          if (colColors[i] != null && ansColors[j] != null
              && colColors[i] === ansColors[j]) {
            cows++;
            colColors[i] = null;
            ansColors[j] = null;
          }
        }
      }
      return cows;
    }
    score(colColors, ansColors) {
      var copyColColors = this.copyArray(colColors);
      var copyAnsColors = this.copyArray(ansColors);
      var bulls = this.scoreBulls(copyColColors, copyAnsColors);
      var cows = this.scoreCows(copyColColors, copyAnsColors);
      // WriteToPara('Bulls: ' + bulls.toString() + ' Cows: ' + cows.toString());
      return [bulls, cows];
    }
    areAllFilled(colors) {
      for (var i = 0; i < colors.length; i++) {
        if (colors[i] == null) {
          return false;
        } 
      }
      return true;
    }
    hitOrMiss() {
      var col = this.cols[this.activeColIndex];
      var colColors = col.getHoleColors();
      //WriteToPara('colColors');
      //WriteToPara(colColors);
      if (!this.areAllFilled(colColors)) {
        return;
      }
      var ansColors = this.ansColors;
      //WriteToPara('ansColors');
      //WriteToPara(ansColors);
      var scores = this.score(colColors, ansColors);
      var pinsCol = this.pinsCols[this.activeColIndex];
      pinsCol.changePinsColors(scores[0], scores[1]);
      this.activeColIndex++;
      return scores[0];
    }
    recolorAnsCol() {
      this.ansCol.changeHoleColors(this.ansColors);
    }
    onClickEvent(x, y) {
      var bulls = this.hitOrMiss();
      if (this.cols.length == this.activeColIndex || bulls == 4) {
        //WriteToPara('DONE');
        this.recolorAnsCol();
        return;
      }
      if (this.pickActiveHole(x, y)) {
        return;
      }
      if (this.pickActivePinsCol(x, y)) {
        var bulls = this.hitOrMiss();
        if (this.cols.length == this.activeColIndex || bulls == 4) {
          this.recolorAnsCol();
        }
        return;
      }
      if (this.colorsPicker.onClickEvent(x, y)) {
        return;
      }
    }
    run() {}
  }
  class Board {
    constructor() {
      this.x = 30;
      this.y = 30;
      this.width = 600;
      this.height = 270;
      this.cols = [];
      this.ansCol = null;
      this.pinsCols = [];
      this.colorsPicker = null;
      this.logo = null;
      this.codeMaker = null;
      this.codeBreaker = null;
    }
    toString() {
      return '[object Board]';
    }
    drawBase() {
      ctx.beginPath();
      ctx.fillStyle = "lightblack";
      ctx.fillRect(this.x, this.y, this.width, this.height);
      ctx.closePath();
      ctx.stroke();
    }
    drawCols() {
      const numCols = 10;
      for (var i = 0; i < numCols; i++) {
        const x = this.x + 60 + i*45;
        const y = this.y + 20;
        const col = new Column(x, y, 'LightSlateGray');
        col.draw();
        this.cols.push(col);
      }
    }
    drawAnsCol() {
      const x = this.x + 20 + 11.5*45;
      const y = this.y + 20;
      const col = new Column(x, y, 'Sienna');
      col.draw();
      this.ansCol = col;
    }
    drawPinsColumns() {
      const numCols = 10;
      for (var i = 0; i < numCols; i++) {
        const x = this.x + 60 + i*45;
        const y = this.y + 170;
        var col = new PinsColumn(x, y, 'yellow');
        col.draw();
        this.pinsCols.push(col);
      }
    }
    drawColorPicker() {
      var colorsPicker = new ColorsPicker(this.x, this.y);
      colorsPicker.draw();
      this.colorsPicker = colorsPicker;
    }
    drawLogo() {
      var logo = new Logo(this.x, this.y);
      logo.draw();
      this.logo = logo;
    }
    createCodeMaker() {
      var codeMaker = new CodeMaker(this.ansCol, this.colorsPicker);
      codeMaker.run();
      this.codeMaker = codeMaker;
    }
    createCodeBreaker() {
      var codeBreaker = new CodeBreaker(
        this.cols, this.ansCol, this.pinsCols,
        this.colorsPicker, this.codeMaker.colors);
      codeBreaker.run();
      this.codeBreaker = codeBreaker;
    }
    onClickEvent(event) {
      //WriteToPara(event);
      var x = event.pageX;
      var y = event.pageY;
      //WriteToPara('x ' + x.toString() + ' y ' + y.toString());
      //WriteToPara(this);
      //WriteToPara(this.codeBreaker);
      this.codeBreaker.onClickEvent(x, y);
    }
    draw() {
      this.drawBase();
      this.drawCols();
      this.drawAnsCol();
      this.drawPinsColumns();
      this.drawColorPicker();
      this.drawLogo();
      //WriteToPara(this.colorsPicker);
    }
    run() {
      this.createCodeMaker();
      this.createCodeBreaker();
    }
  }

  var board = new Board();
  board.draw();
  board.run();

  function onClickEvent(event) {
    //WriteToPara(this);
    board.onClickEvent(event);
  }

  canvas.addEventListener('click', onClickEvent, false);
</script>
      
</body> 
</html>
```
